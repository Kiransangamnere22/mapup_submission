# Question 1: Distance Matrix Calculation

```python
def calculate_distance_matrix(df):
    distance_matrix = df.pivot(index='id_start', columns='id_end', values='distance').fillna(0)
    distance_matrix = distance_matrix + distance_matrix.T.fillna(0)
    distance_matrix = distance_matrix.cumsum(axis=1)
    distance_matrix.values[[range(len(distance_matrix))]*2] = 0
    return distance_matrix
```

```python
import pandas as pd

def calculate_distance_matrix(dataset_path):
    # Load the dataset
    df = pd.read_csv('C:/Users/Asus/Downloads/MapUp-Data-Assessment-F-main/MapUp-Data-Assessment-F-main/datasets/dataset-3.csv')

    # Initialize an empty distance matrix
    unique_ids = sorted(pd.concat([df['id_start'], df['id_end']]).unique())
    distance_matrix = pd.DataFrame(index=unique_ids, columns=unique_ids, dtype=float)

    # Fill in the distance matrix with cumulative distances
    for id1 in unique_ids:
        for id2 in unique_ids:
            if id1 == id2:
                distance_matrix.loc[id1, id2] = 0.0
            else:
                # Find distances from id1 to id2 and id2 to id1
                distances1 = df.loc[(df['id_start'] == id1) & (df['id_end'] == id2), 'distance'].sum()
                distances2 = df.loc[(df['id_start'] == id2) & (df['id_end'] == id1), 'distance'].sum()

                # Set the cumulative distance in both directions
                distance_matrix.loc[id1, id2] = distances1 + distances2
                distance_matrix.loc[id2, id1] = distances1 + distances2

    return distance_matrix

# Example usage
dataset_path = 'C:/Users/Asus/Downloads/MapUp-Data-Assessment-F-main/MapUp-Data-Assessment-F-main/datasets/dataset-3.csv'
result_distance_matrix = calculate_distance_matrix(dataset_path)
print(result_distance_matrix)
```

# Question 2: Unroll Distance Matrix

```python


def unroll_distance_matrix(distance_matrix):
 
    distance_df = distance_matrix.reset_index()

    distance_df = pd.melt(distance_df, id_vars=distance_df.columns[0], var_name='id_end', value_name='distance')

    distance_df = distance_df[distance_df.iloc[:, 0] != distance_df['id_end']]

    return distance_df

result_unrolled_distance = unroll_distance_matrix(result_distance_matrix)
print(result_unrolled_distance)

```

# Question 3: Finding IDs within Percentage Threshold

```python
def find_ids_within_ten_percentage_threshold(distance_df, reference_value):
    reference_avg_distance = distance_df[distance_df['id_start'] == reference_value]['distance'].mean()
    lower_bound = reference_avg_distance - (reference_avg_distance * 0.1)
    upper_bound = reference_avg_distance + (reference_avg_distance * 0.1)
    result_df = distance_df[(distance_df['id_start'] != reference_value) & 
                             (distance_df['distance'] >= lower_bound) &
                             (distance_df['distance'] <= upper_bound)]
    result_ids = sorted(result_df['id_start'].unique())

    return result_ids
```

```python
reference_value = 1001470  
result_within_threshold = find_ids_within_ten_percentage_threshold(result_unrolled_distance, reference_value)
print(result_within_threshold)
```

# Question 4: Calculate Toll Rate

```python
def calculate_toll_rate(distance_df):
    # Define rate coefficients for each vehicle type
    rate_coefficients = {'moto': 0.8, 'car': 1.2, 'rv': 1.5, 'bus': 2.2, 'truck': 3.6}

    # Loop through each vehicle type and calculate toll rates
    for vehicle_type, rate_coefficient in rate_coefficients.items():
        distance_df[vehicle_type] = distance_df['distance'] * rate_coefficient

    return distance_df

# Example usage for Question 4
# Assuming result_unrolled_distance is the DataFrame obtained from Question 2
result_with_toll_rates = calculate_toll_rate(result_unrolled_distance)
print(result_with_toll_rates)

```

# Question 5: Calculate Time-Based Toll Rates

```python
import pandas as pd
import datetime

def calculate_time_based_toll_rates(distance_df):
    # Define time ranges and discount factors
    weekday_time_ranges = [
        (datetime.time(0, 0, 0), datetime.time(10, 0, 0)),
        (datetime.time(10, 0, 0), datetime.time(18, 0, 0)),
        (datetime.time(18, 0, 0), datetime.time(23, 59, 59))
    ]

    weekend_time_ranges = [
        (datetime.time(0, 0, 0), datetime.time(23, 59, 59))
    ]

    weekday_discount_factors = [0.8, 1.2, 0.8]
    weekend_discount_factor = 0.7

    # Create an empty DataFrame to store the result
    result_df = pd.DataFrame()

    # Iterate through each time range and apply discount factors
    for start_day in range(1, 8):  # 1 corresponds to Monday
        for start_time, end_time in weekday_time_ranges:
            for id_start, id_end in distance_df[['id_start', 'id_end']].drop_duplicates().itertuples(index=False):
                # Filter rows based on id_start, id_end, day, and time range
                mask = (
                    (distance_df['id_start'] == id_start) &
                    (distance_df['id_end'] == id_end) &
                    (distance_df['day'] == start_day) &
                    (distance_df['time'] >= start_time) &
                    (distance_df['time'] <= end_time)
                )

                # Apply discount factors
                if start_day <= 5:  # Weekday
                    discount_factor = weekday_discount_factors[weekday_time_ranges.index((start_time, end_time))]
                else:  # Weekend
                    discount_factor = weekend_discount_factor

                # Calculate discounted toll rates
                for vehicle_type in ['moto', 'car', 'rv', 'bus', 'truck']:
                    column_name = f'{vehicle_type}_{start_day}_{start_time}_{end_time}'
                    result_df[column_name] = distance_df[mask][vehicle_type] * discount_factor

    return result_df
```

```python
result_with_time_based_toll_rates = calculate_time_based_toll_rates(result_unrolled_distance)
print(result_with_time_based_toll_rates)
```
